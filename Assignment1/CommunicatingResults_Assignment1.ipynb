## Introduction

Soft Brexit, or hard Brexit? The question that has undoubtedly dominated European news headlines during the last months is about to be answered on October 31st the latest. In a chaotic run-up to the Brexit deadline, current prime minister Boris Johnson is facing increasing backlash against his hard Brexit stance. Surprisingly though, the labour party (Britain's currently biggest opposition party) has not been able to profit from Johnson's struggles. Having been tied with the conservatives back in Mid-July, recent polls suggest that Labour has since fallen behind the Conservative party again, while other parties, such as the Liberal Democrats, are gaining popularity (Politico, n.d.).  
   
Labour's downward trend might well be due to Jeremy Corbyn's ambigious Brexit strategy, which is lacking a clear position and instead advocating for a new referendum should the current government fail (Edgington, 2019). Given that the current circumstances present a unique opportunity to convince new voters, assessing the success of Jeremy Corbyn's Brexit strategy is crucial for future strategic considerations of not only himself, but the labour party as a whole. One way to assess public opinion towards Jeremy Corbyn’s Brexit strategy is analysing the current Twitter discourse.   
   
While doing so comes with a range of limitations that will be elaborated upon later on in this notebook, Twitter analyses also hold certain advantages over traditional methods. For example, they allow researchers to capture expressive behaviour in a natural setting (Russell Neuman, Guggenheim, Mo Jang & Bae, 2014), and allow for more nuance and realistic expressions compared to opinion surveys with pre-made attitude statements. To gauge the public's evaluation of Jeremy Corbyn's Brexit strategy by investigating temporal changes in natural, manifest behaviour on a large scale with limited resources, the following research question is posed:   
   
*RQ1: How does Jeremy Corbyn’s Brexit strategy resonate with the public (on Twitter) as compared to Boris Johnson's?*   
   
When answering this question, it is important to consider changes over time, as well as the unique events that might have shaped and affected public discourse. Specifically, both the labour party and the conservative party held a party conference recently. To account for the latter, a second research question is posed:

*RQ2: What effects did the Labour party convention and Conservative Party Convention have on the public discourse around Jeremy Corbyn’s Brexit strategy?*   
   
The answers to those research questions should not only prove useful for the labour party's strategic considerations, but will also provide insights into the dynamics of public opinion within an increasingly polarised country in a period of inner crisis.

## Hypotheses and Sub-Research Questions

The two reearch questions posed above, will be answered through an explorative analysis that examines the development of evaluation of tweets that mention Jeremy Corbyn over time. In addition to that, two hypotheses test the effect that two major events, namely the party conferences of both Labour and the Conservatives, had on this overall development.   
Naturally, all political parties have a certain set of ideas that they hope to promote and eventually be elected for. To do so however, they they first need to attract the attention of the public. This process is best captured by Cobb, Ross, and Ross (1976), who, within their agenda-building theory, propose a mobilsation model that assumes that political actors aim to move their issues into the public sphere to gain support and legitimation for political change. While the model captures the reasoning for mobilisation efforts by political actors, it makes no predictions about whether or not they are successful. In the context of this analysis however, it seems logical to assume that the Labour party used their party conference as a means to attract increasing attention and support for, not only their general policies, but specifically Jeremy Corbyn's Brexit-strategy. It then follows naturally, that the Conservative party tried to achieve the same with their party conference, thereby taking the spotlight away from Jeremy Corbyn and potentially criticising him too. The following hypotheses aim to test these assmuptions:   

*H1a: The Labour Party Convention increased the evaluation of Jeremy Corbyn.*   
   
*H1b: The Conservative Party Convention decreased the evaluation of Jeremy Corbyn.*

To answer these hypotheses, the construct *"Evaluation of Jeremy Corbyn"* is operationalised as the overall sentiment of tweets that mention Jeremy Corbyn. The hypotheses will be tested through two ANOVAs with three time periods (pre-conference, during-conference, and post-conference) as the independent, and overall sentiment as the dependent variable.   
   
In addition to the effect that the formal party events had on the public discourse, a second hypothesis will examine Jeremy Corbyn's agenda-building power (in terms of its public reception) on a more individual level. Specifically, it will examine whether or not Jeremy Corbyn's (official) Twitter activity is causally related to his evaluations in such a way that increased activity causes more positive sentiment.    
This hypothesis is inspired by several recent agenda-setting studies. For example, Skogerbø and Krumsvik (2015) found that local politicians are increasinly active on social media during election periods, even though these increases in activity do not appear to spill over to traditional news sources in terms of agenda setting. Examining agenda-setting in the context of the Democratic presidential primaries, Conway, Kenski, and Wang (2015) found a symbiotic relationship between news coverage and Twitter activity. Taken together, these findings indicate that Jeremy Corbyn should be more active on Twitter during the heated Brexit debate, as it is such an important matter for the nation as a whole.    
Furthermore, if social media activity has the power to influence traditional newspapers, as indicated by Conway, Kenski, and Wang (2015), it does not seem farfetched to assume that it has even stronger effects in terms of agenda-setting and its public reception in its native social media environment. In other words, Jeremy Corbyn might well be able to influence the public discourse on Twitter through his own participation.   
While it makes little sense to investigate agenda-setting effects in such a dominant topic, it is interesting to see whether Jeremy Corbyn can at least somewhat steer his own evaluation through his Twitter activity. Assuming that he can indeed do so, the following hypothesis is formulated:


*H2: Jeremy Corbyn's Twitter activity is causally related to his evaluation in such a way that increased activity leads to a better evaluation of him.*   
   
The two hypotheses above are interesting and relevant in and of itself, but not necessarily very helpful to guide Jeremy Corbyn's future Twitter activity. For this reason, a third hypothesis is posed that examines whether the emotionality or the overall sentiment of a tweet influences a tweets' reach (operationalised as its total number of retweets). The hypothesis reads as follows:

*H3: The reach of a tweet is predicted by a) its emotionality and b) its overall sentiment.*

Since this project concerns the Twitter discourse around Jeremy Corbyn, not all tweets are necessary for all of the analysis. Hence, in the following, the function "word_present" is created and then used to add a column that indicates whether a tweet mentioned Jeremy Corbyn by his last name (1) or not (0). The function was taken from the "useful functions" notebook on the Digital Analytics github page.

```python
def word_present(text, query):
    import re
    text = str(text).lower()
    query = str(query).lower()
    
    tokens = re.findall(r"[\w']+|[.,!?;$@#]", text)
    if query in tokens:
        return 1
    return 0
```

```python
tweets['corbyn_mentioned'] = tweets['text'].apply(word_present, args=('corbyn',))
```

```python
tweets['johnson_mentioned'] = tweets['text'].apply(word_present, args=('johnson',))
```

## Testing H2: Time Series
H2 assumed that Jeremy Corbyn's Twitter activity is causally related to his evaluation in such a way that increased activity leads to a better evaluation of him. This will be tested by means of a time series analysis in the following. Time series analyses are especially useful to test agenda-setting effects (Russell Neumann, et al., 2014), as they allow researchers to examine how one variable affects another one over time and infer causality. Since H2 is inspired by agenda-setting effects and does indeed examine whether Jeremy Corbyn's tweet frequency on a given day affects tweet sentiment on the following day, time series is the optimal method to test this hypothesis.

### Stationarity

```python
sm.tsa.stattools.adfuller(ts_data['overall_sent_corbyn'])
```

```python
sm.tsa.stattools.adfuller(ts_data['no_of_tweets'])
```

With a p-value below .05, the 'overall_sent' series appears to be stationary. The series for 'no_of_tweets' however is not. Thus, we need to difference. To do so, a new dataframe is created:

```python
ts_data_diff=ts_data
ts_data_diff['overall_sent_corbyn']=ts_data_diff['overall_sent_corbyn'].diff()
ts_data_diff['no_of_tweets']=ts_data_diff['no_of_tweets'].diff()
```

```python
ts_data_diff.head()
```

```python
sm.tsa.stattools.adfuller(ts_data_diff.ix[1:]['overall_sent_corbyn'])
```

```python
sm.tsa.stattools.adfuller(ts_data_diff.ix[1:]['no_of_tweets'])
```

As can be seen in the output abve, both variables are now stationary.

In the following a VAR model is estimated. The first line in the dataset is left out, as it contains NaN after the differencing. For simplicity, only one lag is used. However, this also might make some sense, as tweets are usually unlikely to have long-lasting effects.

```python
dfsub = ts_data_diff.ix[1:][['overall_sent_corbyn','no_of_tweets']]
model = sm.tsa.VAR(dfsub)
results = model.fit(1)
results.summary()
```

```python
print(results.test_causality('overall_sent_corbyn',['no_of_tweets']))
```

```python
print(results.test_causality('no_of_tweets',['no_of_tweets']))
```

Based on the output, as well as the causality tests, H2 has to be rejected. Increased Twitter activity by Jeremy Corbyn does not cause any changes in sentiment. Neither do changes in sentiment affect Jeremy Corbyn's tweeting frequency. The only aspect that does seem to predict Jeremy Corbyn's tweet frequency is the number of tweets that his official account published the previous day.
   
in the following, these effects will be visualised (note that only the number of tweets effect on the number of tweets was significant).

```python
results.plot()
```

```python
irf = results.irf(1)
irf.plot()
```

Zooming in on the effect that no_of_tweets has on no_of_tweets, the graph shows a strong negative effect within 1 time lag where an increase of 1 tweet results in a decrease of ~.30 tweets the following day. This does indeed align with the earlier results graph which already indicated that a day with many tweets is usually followed by a day with quite a few tweets less.    
   
Next, I compute the *forecast error variance* (FEV), which is a good indicator to show how much of the movement of
one variable is attributed by shocks of its own past and to what extent by shocks of the other variables. The method can show how much variation of one variable can be explained by its own past or by another variable.

```python
fevd = results.fevd(1)
fevd.summary()
```

The results of the FEV indicate that after one day, 2.8% of variation in the number of tweets can be explained by the overall sentiment. However, remember that this effect was not significant.
   
To ensure that the constructed VAR model is robust several tests need to be conducted. The first test is the Portmonteau (Q) test. Its null hypothesis assumes the residuals to be white noise, meaning that there is no autocorrelation in the time series. If autocorrelation is present in the data, that means that the time series is highly correlated with a lagged version of itself.   
The comand "results.residuals" returns residuals values needed in robustness checks and is used for the following tests. Note that the first number of the next two outputs contains the test statistics, and the second number the p value.

```python
sm.stats.diagnostic.acorr_ljungbox(results.resid['no_of_tweets'],lags=1)
```

```python
sm.stats.diagnostic.acorr_ljungbox(results.resid['overall_sent_corbyn'],lags=1)
```

The tests for autocorrelation indicate that residuals morror white noise (at 1st lang, no_of_tweets: 0.02, ns; overall_sent: 0.72, ns).   
   
Next, the data is checked for heteroscedasticity. This is done by performing the Portmonteau (Q) test again, but now by looking at the squared residuals of the series. 

```python
sres_no_of_tweets=[]
```

```python
for line in results.resid['no_of_tweets']:
    r_s=line*line
    sres_no_of_tweets.append(r_s)
```

```python
sm.stats.diagnostic.acorr_ljungbox(sres_no_of_tweets,lags=1)
```

```python
sres_overall_sent=[]
for line in results.resid['overall_sent_corbyn']:
    r_s=line*line
    #print(r_s)
    sres_overall_sent.append(r_s)
```

```python
sm.stats.diagnostic.acorr_ljungbox(sres_overall_sent,lags=1)
```

Based on the output, we can not reject the null hypothesis, which indicates that there might be heterodasticity present in the data.   
   
The final test - the VAR stability test - checks the eigenvalue stability condition after estimating the VAR model. The modulus of each eigenvalue has to be less than 1 so that the VAR model can be considered stable. If the VAR model is considered stable, we can interpret the CIRF and FEVD as outlined above.

```python
smtsa.vector_ar.var_model.VARProcess.is_stable(results)
```

Despite its insignificance and concerns over herodasticity, the returned value 'True' allows us to coclude that the model is stable. This means that the FEVD interpretations made earlier are correct.

As can be seen in this visualisation as well as in the results of the time series analysis, Jeremy Corbyn's tweet frequency had no significant effect on his evaluation on Twitter. While the frequency didn't matter, it might still be that the content did. However, this was not accounted for in this analysis.   
   
Overall, it has to be concluded that Jeremy Corbyn's Brexit strategy was evaluated worse over time by the public on Twitter and that neither the party conference, nor his own official twitter activity was able to turn this trend around. This suggests that Jeremy Corbyn should overthink either his Brexit-position, which, as noted earlier, is rather ambiguous (Edgington, 2019), or his and his party's online communication of it.   
   
H3 aimed at finding some meaningful suggestions for how Jeremy Corbyn might improve his tweets to gain additional reach (retweets). However, the hypothesised elements of a tweet, namely emotionality and sentiment, dd nt have a significant effect on retweets in the overall dataset. This means that other features, possibly length, topic, or specfic hashtags might be more helpful. However, future analyses will be needed in order to test whether these elements actually do have an effect.

## Limitations and Next Steps

While analysing the Twitter discourse around Jeremy Corbyn has the advantage of being a cost-effective measure of manifest behaviour, it doesn't come without limitations. On a general level, it is important to note that a bigger dataset in and of itself doesn't make the results more meaningful (Boyd & Crawford, 2012). For example, the current analysis relied on a large dataset of tweets that contained the word "Brexit" and/or one or both of two associated hashtags (#Remainers, #Brexitiers). While this sample naturally encompasses various tweets that refer to Jeremy Corbyn's Brexit strategy, it might have been too broad. That is, because it can't necessarily be expected that every tweet that mentions Jeremy Corbyn also presents an overall evaluation of him. In fact, some of the most common hashtags (e.g. #leaveoct31) point to this being the case.   
Similarly, the sentiment analysis develloped by Thelwall, et al.(2010) only distinguished between positive, negative and neutral words. This does not take into account nuances in how positive or negative a word might be, nor does it pay attention to the specific context in which some words might take on different meanings.   
   
As for the data-preparation, it has to be noted that no distinction was made between tweets that mentioned only Jeremy Corbyn and tweets that mentioned him together with prime minister Boris Johnson. While the latter ategory only made for 10% of the total sample, the fact that those tweets were more negative on average might still have skewed the later analysis. Furthermore, the assumption that a tweet that mentions not only Jeremy Corbyn but also Boris Johnson still constitutes an evaluation of the former might in fact be a little too far fetched.   
   
The abovementioned limitations might somewhat call into question the value of the results gained in the analysis. As for the ANOVA, it is also worth pointing out that a) the time periods post-LC and pre-CC were overlapping, b) the four-day period was chosen rather arbitrarily, and c) it doesn't take into account other events that happened at the time. Regarding the ANOVA-results, especially limitations a) and c) are concerning. That is, because the time period after the labour party conference was rather short and with the tory party conference coming up, positive effects from it might not have lasted for four full days. Furthermore, events such as the supreme court ruling that the suspension of parliament was unlawful on the 24th of September, as well as personal scandals of Boris Johnson might have affected the Twitter discourse around Brexit in ways that this analysis does not account for. In other words, full causality can not be deduced from the findings of the two ANOVAs as other confoundig variables might have caused the differences.       
Another aspect concerning the hypothesis testing that is noteworthy, is that the time series analysis was performed with only one lag. While it can be assumed that Jeremy Corbyn's daily tweet-frequency does not have long-lasting effects on sentiment, limiting the lags to just one from the very beginning might still be a little too much. Then again, no meaningful effects of one variable on the other were found, so having more lags might not necessarily make sense.   
Lastly, regarding H3, it needs to be noted that analysing the effect that emotionality and sentiment had on retweets in the whole sample might be problematic insofar, as th popularity of different accounts could have confounded possible effects. Hence, the analysis should be repeated for Jeremy Corbyn's personal set of tweets.

### Next steps
The current analysis would profit from both, methodological improvements and differentiated follow-up studies. As for improvements, it would be worth comparing different sentiment analysis tools, to evaluate which one fits the context best and is able to provide most nuance. Thereby, the sentiment analysis could move beyond its currently rather crude measurement. Similarly, it would be worthwhile to re-do the ANOVAs with different (probably shorter) time frames to avoid overlapping effects. Moreover, to gain a better understanding of what is happening, it might be necessary to consider more important events than just the two party conferences, or, alternatively, zoom in on for example Jeremy Corbyn's speech and test for direct effects that this had.    
   
From an academic standpoint, it would also be interesting to repeat the analysis with "emotionality" as the dependent variable, since, unlike with sentiment, polarisation does not hide its variation.   
   
The most important question of course is, how the labour party can profit from this analysis. First of all, survey and focus group data might help to evaluate the labour party conference, which clearly did not have the expected effects (unlike the conservative party conference).   
Another interesting notion is, that Twitter is more useful for discrediting opponents than it is for gaining support and positive feedback of ones' own policies. The fact that the conservative party conference had the expected negative effect on Jeremy Corbyn's evaluation points in this direction. Naturally, this claim would need to be tested by seeing whether the labour party conference had a negative effect on Boris Johnson's sentiment, but if it would prove to be true, this insight might help Jeremy Corbyn and his advisors to use Twitter more effectively.
It might also be worth to consider using sentiment topic recognition (STR) in order to find the most representative topics discussed for each sentiment, as suggested by Mostafa (2013). This might help provide more detailled insights into what Twitter users are most concerend with. Furthermore, as noted earlier, future analysis could investigate which features of a tweet improve its reach. This way, Jeremy Corbyn could at least ensure that he reaches more people on Twitter.

## Ethical, Privacy and Normative Considerations

Gathering data for Twitter analyses is comparably easy and cost-efficient. However, an aspect that is often forgotten are ethical considerations, which receive a lot more attention in traditional academic methods such as experiments (Boyd & Craword, 2012). When it comes to this analysis, several potential ethical and privacy issues come to mind.   
   
First of all, it is noteworthy that there is an important difference between "being in public (i.e. sitting in a
park) and being public (i.e. actively courting attention)" (Boyd & Marwick 2011, as cited in Boyd & Crawford, 2012). Applied to Twitter, this analogy means that while tweets are public by default and most Twitter users can be expected to be aware of that, this fact in itself doesn't mean that those users like the idea of having their tweets analysed. More importantly, there is no way for those users to know that their content is being analysed (although in an anonymised and aggregated form), nor do they have an opportunity to have their tweets removed from the dataset. The latter issue does not only seem unfair, but actually contradicts with the right to be forgotten.   
   
Another issue is the amount of data that was collected in the first place. According to GDPR rules, one is only allowed to collect data when there is a plan on how to use it. With this dataset, this was not the case as many variables in the original dataset were not used in the analysis. While the GDPR moght not apply to Twitter data, this issue nonetheless goes against what one might describe as good practice.   
   
That said, compared to commercial data usage by various companies, this analysis is rather harmless. That is, because the data was gathered from Jeremy Corbyn's public account, as well as from anaonymised Twitter users. However, as soon as the labour party would try to gather more personal data from external data brokers (data that must surely be available as Twitter users might be tracked beyond just Twitter), it would run the risk of invading users privacy and contributing to a questionable business model in which data brokers operate largely in the dark and face little regulatory oversight (Martin, 2015).
